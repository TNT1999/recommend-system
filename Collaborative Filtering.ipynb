```python
import requests
import pandas as pd
import numpy as np

url = 'http://localhost:8000/api/allView'
r = requests.get(url)
data = r.json()
viewCount = pd.DataFrame(data).dropna()
```

```python
viewCount.rename(columns = {'count':'view_count'}, inplace = True)
viewCount
```

```python
group_by_user = viewCount.groupby('userId')
pd.DataFrame(group_by_user.count()["productId"])

# group_by_user["number_Product_view"] = pd.DataFrame(viewCount.groupby('userId').count()["productId"])
# group_by_user
```

```python
group_by_product = viewCount.groupby('productId')
pd.DataFrame(group_by_product.count()["userId"])
```

```python
user_product = viewCount.pivot_table(index="productId",columns="userId",values='view_count').fillna(0)
user_product
```

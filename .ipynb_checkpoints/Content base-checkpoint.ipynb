```python
import requests
import pandas as pd

url = 'http://localhost:8000/api/allProduct'
r = requests.get(url)
data = r.json()
products = pd.DataFrame(data)
print(products)
```

```python
# products['brand'].value_counts()
products.head(1)["attrs"]
```

```python
import re

def clean_name(name):
#     title = re.sub("[^a-zA-Z0-9 ]", "", title)
    return name

def clean_attrs(attr):
    print(attr.value)
    return attr.value
```

```python
products["name"] = products["name"].apply(clean_name)
products["name"] = products["name"].str.lower()
products["name"] = products["name"].str.replace("\s+", " ", regex=True)
products = products[products["name"].str.len() > 0]
products["attrs"] = products["attrs"].apply(clean_attrs)
```

```python
products
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(ngram_range=(1,3))

tfidf = vectorizer.fit_transform(products["name"])
```

```python
a = vectorizer.get_feature_names_out()
print(a,tfidf.shape)
```
